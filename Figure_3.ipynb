```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

start = 90+144+144+144+144+144
end = 120+144+144+144+144+144

x_axis = np.linspace(15, 20, num=30)

# Plotting the first graph: 'Weather-based prediction'
plt.plot(x_axis, wb_df['pred'][start:end], label='Weather-based prediction', color='red')

# Plotting the second graph: 'Non-weather-based prediction'
plt.plot(x_axis, nwb_df['pred'][start:end], label='Non-weather-based prediction', color='blue')

# Plotting the third graph: 'Actual Power Output'
plt.plot(x_axis, wb_df['y_test'][start:end], label='Actual Power Output', color='green')

# Customize the axes labels
plt.xlabel('Hour of the day')
plt.ylabel('Power generated in that hour (kW)')


# Display the legend
plt.legend()

# Show the overlaid graphs
plt.show()
```
